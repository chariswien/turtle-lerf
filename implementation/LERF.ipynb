```python
import torch
import numpy as np
import open3d
from pathlib import Path

from nerfstudio.utils.eval_utils import eval_setup
from nerfstudio.cameras.rays import RayBundle
```

```python
# Model laden 

config_path = Path("/app/datasets/test_1509/output/input/lerf/2025-09-15_160120/config.yml")

# Aus https://github.com/nerfstudio-project/nerfstudio/blob/main/nerfstudio/utils/eval_utils.py
_, pipeline, _, _ = eval_setup(
    config_path,
    eval_num_rays_per_chunk=None,
    test_mode='test'
)

model = pipeline.model

print("+ + + + + succes + + + + +")
```

```python
#Pointcloud laden und entsprechend dataparser transformieren


#ponitcloud laden und als numpy array speichern
path_pointcloud = "/app/exports/pcd/point_cloud.ply"
pointcloud = open3d.io.read_point_cloud(path_pointcloud) #aus https://www.open3d.org/docs/latest/python_api/open3d.io.read_point_cloud.html
pointcloud_array = np.asarray(pointcloud.points) #aus https://www.open3d.org/docs/0.8.0/tutorial/Basic/working_with_numpy.html

print("+ + + Succes:pointcloud geladen und als array gespeichert + + +")
print(f"Es sind {len(pointcloud_array)} punkte gespeichert")


#tf und scale aus dem dataparser lesen
dataparser = pipeline.datamanager.dataparser #aus https://docs.nerf.studio/reference/api/data/dataparsers.html
dataparser_outputs = dataparser.get_dataparser_outputs(split='train')
transform_matrix_dataparser = dataparser_outputs.dataparser_transform
scale_factor_dataparser = dataparser_outputs.dataparser_scale

print(f"dataparser tf:\n {transform_matrix_dataparser}")
print(f"dataparser scale: {scale_factor_dataparser}")
```

```python
# TF basierend auf nerf-transform (test)

DATATRANSFORM_APPLY = False  #nicht notwendig --> auf Flase lassen

if DATATRANSFORM_APPLY == True:
    print("Dataparser tf wird angewendet")

    #World-KS in hom 4x4-Matrix 
    #points_hom = np.hstack([pointcloud_array, np.ones((len(pointcloud_array), 1))])
    
    #dataparser transform matrix auf 4x4 bringen
    if transform_matrix_dataparser.shape[0] == 3:
        # [0, 0, 0, 1] ergänzen
        transform_matrix_4x4 = np.vstack([transform_matrix_dataparser, [0, 0, 0, 1]])
    else:
        transform_matrix_4x4 = transform_matrix_dataparser
        
    #ausgabe
    print(f"Transform 4x4: {transform_matrix_4x4.shape}")

    translation = transform_matrix_4x4[:3, 3]

    points_model = (pointcloud_array - translation) / scale_factor_dataparser

    print(f"Transform angewendet:")
    print(f"  Translation: {translation}")
    print(f"  Scale: {scale_factor_dataparser}")

else:
    print("Keine Dataparser tf - Original Koordinaten behalten")
    # punkte unverändert
    points_model = pointcloud_array.copy()

# Ausgabe
print(f"model points shape: {points_model.shape}")
print(f"world coordinate range: {pointcloud_array.min(axis=0)} to {pointcloud_array.max(axis=0)}")
print(f"Model coordinate range: {points_model.min(axis=0)} to {points_model.max(axis=0)}")
```

```python
#query setzen
query = "green plant in a pot"
canonical_phrases = ["object", "things", "stuff", "texture"] #LERF paper 3.5

# pos_embeds: normalized embeddings der query
# neg_embeds: normalized embeddings der canonical phrases (kommen aus dem model; repo s.u.)

print(f"Processing query: '{query}'")
print("="*60)

model.image_encoder.set_positives([query]) # aus repo https://github.com/kerrj/lerf/blob/main/lerf/encoders/clip_encoder.py

print("\nCanonical phrases (from model config (LERF repo):") #Lerf paper: Kap. 3.5 // repo: https://github.com/kerrj/lerf/blob/db08d578038d884542688511bd9ad7b489a65673/lerf/encoders/clip_encoder.py
if hasattr(model.image_encoder, 'negatives'):
    for neg in model.image_encoder.negatives:
        print(f"  - {neg}")
    print(f"Negative embeddings shape: {model.image_encoder.neg_embeds.shape}")
else:
    print("  No negatives found in model")


```

```python
def get_clip_features_from_positions(model, positions, scale=0.5):

    # entspricht get_outputs aus https://github.com/kerrj/lerf/blob/main/lerf/lerf_field.py

    with torch.no_grad():

        positions = model.lerf_field.spatial_distortion(positions)
        positions = (positions + 2.0) / 4.0
        
        xs = [e(positions.view(-1, 3)) for e in model.lerf_field.clip_encs]
        
        x = torch.concat(xs, dim=-1)
        
        clip_scales = torch.ones((len(positions), 1), device=positions.device) * scale
        clip_pass = model.lerf_field.clip_net(torch.cat([x, clip_scales.view(-1, 1)], dim=-1))
        
        # normalisieren
        clip_features = clip_pass / clip_pass.norm(dim=-1, keepdim=True)
    
    return clip_features


print("function created based on lerf implementation")
```

```python
scale_lerf = 0.2 # change for other scales

# Then use that optimal scale for processing
batch_size = 10000
all_relevances = []

print(f"\nProcessing {len(points_model)} points for query: '{query}'")
print(f"Batch size: {batch_size}")
print(f"Using scale: {scale_lerf:.2f}") 

for i in range(0, len(points_model), batch_size):
    if i % 50000 == 0 and i > 0:
        print(f"  Progress: {i}/{len(points_model)} points processed")
    
    batch_end = min(i + batch_size, len(points_model))
    batch_points = points_model[i:batch_end]
    positions = torch.from_numpy(batch_points).float().cuda()
    
    # Use optimal_scale instead of fixed scale
    clip_features = get_clip_features_from_positions(model, positions, scale_lerf)
    
    relevancy = model.image_encoder.get_relevancy(
        embed=clip_features,
        positive_id=0
    )
    relevancy_scores = relevancy[:, 0]
    all_relevances.append(relevancy_scores.cpu().numpy())

relevances = np.concatenate(all_relevances)

```

```python
# Set threshold and create filtered PLY
threshold = 0.75
print(f"\nFiltering with threshold: {threshold:.3f}")

# Create mask for points above threshold
relevant_mask = relevances > threshold
relevant_points_world = pointcloud_array[relevant_mask]

print(f"Points selected: {len(relevant_points_world):,} / {len(pointcloud_array):,}")
print(f"Percentage: {100*len(relevant_points_world)/len(pointcloud_array):.2f}%")

# Create and save filtered point cloud
if len(relevant_points_world) > 0:
    import os
    
    # Create output directory
    output_dir = "/app/datasets/test_1509/output/semantics"
    os.makedirs(output_dir, exist_ok=True)
    
    # Create point cloud
    pcd_filtered = open3d.geometry.PointCloud()
    pcd_filtered.points = open3d.utility.Vector3dVector(relevant_points_world)
    
    # Uniform red color for ROS2 compatibility
    colors = np.ones((len(relevant_points_world), 3)) * [1.0, 0.0, 0.0]
    pcd_filtered.colors = open3d.utility.Vector3dVector(colors)
    
    # Generate filename
    threshold_int = int(threshold * 100)
    filename = f"{query.replace(' ', '_')}_th{threshold_int}sc{int(scale_lerf*100)}.ply"
    output_path = os.path.join(output_dir, filename)
    
    # Save
    open3d.io.write_point_cloud(output_path, pcd_filtered)
    
    print(f"\nSaved: {output_path}")
    print(f"Points: {len(relevant_points_world):,}")
    
    # Save metadata
    metadata_path = output_path.replace('.ply', '_metadata.txt')
    with open(metadata_path, 'w') as f:
        f.write(f"Query: {query}\n")
        f.write(f"Scale: {scale_lerf}\n")
        f.write(f"Threshold: {threshold:.6f}\n")
        f.write(f"Points: {len(relevant_points_world)}/{len(pointcloud_array)}\n")
        f.write(f"Percentage: {100*len(relevant_points_world)/len(pointcloud_array):.2f}%\n")
        f.write(f"Relevance range: [{relevances[relevant_mask].min():.3f}, {relevances[relevant_mask].max():.3f}]\n")
    
    print(f"Metadata: {metadata_path}")
else:
    print(f"\nNo points found above threshold {threshold:.3f}")
```
